###Number guessing jackpot

```python
rnmbr=7
for i in range(1,6):
  user_input=(input('Guess a number'))
  if user_input=='x':
    print('Bye Bye!')
    break
  if not user_input.isdigit():
    print('please enter a valid number or x to exit')
    continue




  gnmbr=int(user_input)
  if rnmbr==gnmbr:
    print('you guess it right')
    break
  elif gnmbr<rnmbr:
    print('Too LOw!')
  else:
    print('Too High!')
  if i==5:
    print('Game Over!')
```
