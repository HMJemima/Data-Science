List

```python
#empty list
print([])
#1D list
print([1,2,3])
#2D list
print([[1,2],[2]])
#3D list
print([[[2],[1,2]]])
#Hetregenous
print([1,2,3,'hello'])
#using type conversion
print(list('hello'))
```

```python
#indexing
l=[1,2,3,4,5]
#pos
print(l[0])
#neg
print(l[-1])
r=[[1,2,3],[4,5,6]]
print(r[1][1])
print(r[-1][-2])
#slicing
q=[1,2,3,4,5,6]
print(q[1:3])#last one is not included
print(q[::2])
print(q[::-2])
```

Accessing Items from a List


```python
l=[1,2,3,4]#append
l.append(1)#adds single value
print(l)
```

```python
l=[1,2,3,4,5]
l.append([1,2])
print(l)
```

```python
#extend
l=[1,2,4,5]
l.extend([1,2])
print(l)
```

```python
l=[1,2,3,10]
l.extend('hello')#extend adds multiple values
print(l)
```

```python
#insert
l=[1,2,3,4,5]
l.insert(2,10)#1st one is position and 2nd one is value
print(l)
```

Editing items in a List

```python
j=[1,2,3,4]
j[0]=500
print(j)
j[1:3]=[100,200,300]
print(j)
```

Deleting items from a List

```python
#del
l=[1,2,3,4,4]
del(l[0])
print(l)
del(l[1:3])
print(l)
```

```python
#remove
l=[1,2,3,4]
l.remove(1)
print(l)
```

```python
#pop
l=[1,2,3,4,5]
l.pop()
print(l)
l.pop(1)
print(l)
```

```python
#clear
j=[1,2,3,4,5]
j.clear()
print(j)
```
