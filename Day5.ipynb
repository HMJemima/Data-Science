Sets
A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:
Unordered
- Mutable
- No Duplicates
- Can't contain mutable data types

Creating Set

```python
#empty
s={}
print(s)
print(type(s))#so this is not correct method
s=set()
print(s)
print(type(s))
#1d or 2d
s={1,2,3,4}
print(s)
#s={1,2,3,4,{1,2,3,4}} #2d is not allowed since set can  ot contain any mutable item
#print(s)
#hetro
s={1,2,3,True,'hello'}
print(s) #here all of members is printed except True ,true means 1 the set already contain 1 and it can print duplicate items.
#using type conversion
s=set([1,2,3,4])
print(s)
```

```python
s={1,2,3}
s1={2,1,3}
print(s==s1)
```

#intersection/intersection_update

#intersection/intersection_update


```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.intersection(s2)
s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
#difference/difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 - s2
s1.difference(s2)
s1.difference_update(s2)
print(s1)
print(s2)
```

```python
#symmetric difference/symmetric difference update
s1={1,2,3,4,5}
s2={4,5,6,7,8}

s1.symmetric_difference(s2)
s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

Frozenset
Frozen set is just an immutable version of a Python set object

```python
# create frozenset
fs1 = frozenset([1,2,3])
print(fs1)
```

```python
# what works and what does not
#union,intersection this all operation will work but not editing
# works -> all read functions
# does't work -> write operations
fs1=frozenset([1,2,3])
fs2=frozenset([3,2,4,5])
print(fs1|fs2)
print(fs1&fs2)
print(fs1-fs2)
print(fs1^fs2)
#fs1.add(4)
#2D frozenset
a=frozenset([1,2,3,frozenset([1,2])])
print(a)
```

```python
# using existing dict
distances = {'delhi':1000,'mumbai':2000,'bangalore':3000}
{ i:j*0.62 for (i,j) in distances.items()} #d.item() returns tuples so (i,j) in distances.items()
```

```python
#using zip
days = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]
{i:j for (i,j) in zip(days,temp_C)}
```

```python
#using if condition
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}
{i:j for i,j in  products.items() if j>10}
```
